```python
#Importamos las librerias necesarias

import cv2
import numpy as np

# Función para calcular el área de un polígono utilizando la fórmula de Shoelace
def calcular_area_poligono(puntos):
    n = len(puntos)
    area = 0
    for i in range(n):
        j = (i + 1) % n
        area += puntos[i][0] * puntos[j][1]
        area -= puntos[j][0] * puntos[i][1]
    return abs(area) / 2.0

# Calcular la escala de píxeles a cm
def calcular_escala(width, height, width_cm, height_cm):
    escala_x = width_cm / width
    escala_y = height_cm / height
    return escala_x, escala_y

# Definimos todo lo que queremos hacer cuando dibujamos el poligono
def seleccionar_poligono(event, x, y, flags, param):
    global puntos, seleccionando, imagen_copia
    
    # Si el usuario hace clic con el botón izquierdo
    if event == cv2.EVENT_LBUTTONDOWN:
        puntos.append((x, y))  # Guardar las coordenadas del punto
        seleccionando = True  # Activar el modo de selección
        
    # Si el usuario suelta el botón izquierdo
    elif event == cv2.EVENT_LBUTTONUP and seleccionando:
        # Dibujar un círculo en el punto seleccionado
        cv2.circle(imagen_copia, (x, y), 3, (0, 255, 0), -1)
        cv2.imshow('Mamografía', imagen_copia)
        
        # Si ya hay al menos dos puntos, dibujar una línea entre el punto actual y el anterior
        if len(puntos) >= 2:
            cv2.line(imagen_copia, puntos[-2], puntos[-1], (0, 255, 0), 2)
            cv2.imshow('Mamografía', imagen_copia)
    
    # Si el usuario hace doble clic con el botón izquierdo
    elif event == cv2.EVENT_LBUTTONDBLCLK and len(puntos) >= 3:
        # Dibujar una línea entre el primer y el último punto para cerrar el polígono
        cv2.line(imagen_copia, puntos[-1], puntos[0], (0, 255, 0), 2)
        cv2.imshow('Mamografía', imagen_copia)
        # Calcular el área del polígono
        area_pixel = calcular_area_poligono(puntos)
        print("Área del polígono seleccionado (en píxeles):", area_pixel)

        # Calcular la escala de píxeles a cm, definimos que la imagen es de 15x15 cm
        escala_x, escala_y = calcular_escala(imagen.shape[1], imagen.shape[0], 15, 15)

        # Calcular el área en cm²
        area_cm2 = area_pixel * escala_x * escala_y
        print("Área del polígono seleccionado (en cm²):", area_cm2)

        # Dibujar el área del polígono en la imagen copia
        cv2.putText(imagen_copia, f"Area: {area_cm2:.2f} cm2", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

        # Guardar la imagen con el polígono y el área
        cv2.imwrite('mamografia_con_poligono.jpg', imagen_copia)

        # Mostrar la imagen con el polígono y el área
        cv2.imshow('Mamografía con Polígono', imagen_copia)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

# Cargar la imagen de la mamografía
imagen = cv2.imread('mamografia.PNG')
imagen_copia = imagen.copy()

# Inicializar variables
puntos = []
seleccionando = False

# Crear una ventana y asociar la función de manejo de eventos del mouse
cv2.namedWindow('Mamografía')
cv2.setMouseCallback('Mamografía', seleccionar_poligono)

# Mostrar la imagen
cv2.imshow('Mamografía', imagen)

# Esperar hasta que el usuario presione la tecla 'q' para salir
while cv2.getWindowProperty('Mamografía', cv2.WND_PROP_VISIBLE) >= 1:
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):
        break

# Cuando el usuario presiona 'q', cerrar todas las ventanas
cv2.destroyAllWindows()
```
