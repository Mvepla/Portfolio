```python
import folium
import pandas as pd
```

```python


# Crea un mapa de Folium centrado en una ubicacin geogrfica
m = folium.Map(location=[39.46763046252599, -0.3671284534416607], zoom_start=13)

# Carga tus datos geoespaciales (por ejemplo, GeoJSON o Shapefile)
# En este ejemplo, utilizaremos un GeoJSON de estados de EE. UU.
geo_json_data = r'C:\Users\funny\Desktop\Bootcamp\mesa_trabajo_core\repo_sept_23\proyectos\EDA\Data\distritos.valencia.limpio.choro.geojson'

# Carga tus datos de valores para cada rea geogrfica
# En este ejemplo, utilizaremos datos de poblacin por estado
data = pd.read_csv(r'C:\Users\funny\Desktop\Bootcamp\mesa_trabajo_core\repo_sept_23\proyectos\EDA\notebooks\alquiler_venta_f')

# Crea un mapa de cloropleth
folium.Choropleth(
    geo_data=geo_json_data,
    name='choropleth',
    data=data,
    columns=['distrito', 'precio venta'],
    key_on='feature.properties.nombre',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Precio venta por distrito').add_to(m)

# Agrega un control de capas
folium.LayerControl().add_to(m)

# Guarda el mapa en un archivo HTML
#m.save('cloropleth_map.html')
m
```
