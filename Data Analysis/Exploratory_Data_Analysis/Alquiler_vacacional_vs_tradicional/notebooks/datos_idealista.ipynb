```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter
```

```python
idealista = pd.read_csv(r'C:\Users\funny\Desktop\Bootcamp\mesa_trabajo_core\repo_sept_23\proyectos\EDA\Data\venta_alquiler_idealista.csv',sep=';')
idealista
```

```python
idealista = idealista.drop(columns= ['fecha venta','fecha alquiler'], axis=1)
idealista
```

```python
plt.figure(figsize=(16,8))
sns.barplot(data=idealista, x='Distrito',y='venta €/m2')
plt.xticks(rotation=-30)
plt.show()
```

```python
# Establecer el estilo de Seaborn
sns.set(style="whitegrid")

# Tamaño de la figura
plt.figure(figsize=(12, 8))

# Utilizar una paleta de colores predefinida de Seaborn (Set2)
barplot = sns.barplot(data=idealista, x='Distrito', y='venta €/m2', palette="Set2")

# Rotar los títulos de los distritos en el eje X
plt.xticks(rotation=45)

# Añadir título y etiquetas de ejes
plt.title('Precio por Metro cuadrado para inmuebles a la venta por Distrito')
plt.xlabel('Distrito')
plt.ylabel('€/m2')

# Añadir separadores de miles en el eje Y (si es necesario)
barplot.get_yaxis().set_major_formatter(FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

# Mostrar la trama
plt.show()
```
