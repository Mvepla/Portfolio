```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
alquiler_venta = pd.read_csv(r'C:\Users\funny\Desktop\Bootcamp\mesa_trabajo_core\repo_sept_23\proyectos\EDA\notebooks\alquiler_venta',index_col='Unnamed: 0')
metros_cuadrados= pd.read_csv(r'C:\Users\funny\Desktop\Bootcamp\mesa_trabajo_core\repo_sept_23\proyectos\EDA\notebooks\metros_cuadrados', index_col='Unnamed: 0')
```

```python

alquiler_venta['Distrito'] = alquiler_venta['Distrito'].str.lower()
alquiler_venta =alquiler_venta.rename(columns = {'Distrito':'distrito'})
alquiler_venta.sort_values(by='distrito', inplace=True)
alquiler_venta
```

```python
metros_cuadrados.sort_values(by='distrito', inplace=True)

metros_cuadrados
```

```python
alquiler_venta_f = pd.merge(alquiler_venta, metros_cuadrados)

alquiler_venta_f
```

```python
alquiler_venta_f['precio venta'] = alquiler_venta_f['venta €/m2'] * alquiler_venta_f['metros2']
alquiler_venta_f['precio alquiler'] = alquiler_venta_f['alquiler €/m2'] * alquiler_venta_f['metros2']
alquiler_venta_f
```

```python
alquiler_venta_f.to_csv('alquiler_venta_f')
```
